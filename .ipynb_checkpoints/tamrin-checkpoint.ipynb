```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
img = cv.imread(r'D:\AnacondaProject\img1.jpg')
imnot = cv.bitwise_not(img)
resl = np.maximum(img, imnot)
plt.imshow(img)
plt.show()
plt.close()
plt.imshow(imnot)
plt.show()
plt.close()
plt.imshow(resl)
plt.show()
plt.close()
```
