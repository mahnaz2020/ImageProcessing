```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
img = cv2.imread(r'D:\AnacondaProject\mm.png')
img2 = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img2)
plt.show()
plt.close()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
img = cv2.imread(r'D:\AnacondaProject\mm.png')
img2 = cv2.flip(img,0)

plt.imshow(img2)
plt.show()
plt.close()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
img = cv2.imread(r'D:\AnacondaProject\mm.png')
img2 = cv2.blur(img, (100, 100))
plt.imshow(img2)
plt.show()
plt.close()
```

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread(r'D:\AnacondaProject\mm.PNG')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
plt.close()
```
