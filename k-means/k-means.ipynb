```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = cv.imread(r'D:\AnacondaProject\picture.jpg')#orginal image
plt.imshow(img)
plt.show()
plt.close()
z = img.reshape((-1,3))
z = np.float32(z)
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER,10,1.0)
k1 = 16
ret1,label1,center1 = cv.kmeans(z,k1,None,criteria,10,cv.KMEANS_RANDOM_CENTERS)
center1 = cv.convertScaleAbs(center1)
pic1 = center1[label1.flatten()]
res = pic1.reshape((img.shape))
plt.imshow(res)
plt.show()
plt.close()
```
